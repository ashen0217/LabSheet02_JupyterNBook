```python
movie_data = pd.read_csv("movie_dataset.csv",header=None)
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
records=[]
for x in range(0,7501):
    records.append([str(movie_data.values[x,y])for y in range(0,20)])
```

```python
records
```

```python
association_rules = apriori(
    records,
    min_support=0.01,
    min_confidence=0.5,
    min_lift=1.0,
    max_length=2
)
```

```python
results = list(association_rules)
```

```python
results
```
